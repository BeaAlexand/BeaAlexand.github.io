```python
df_combined = pd.concat([df_2022, df_2023, df_2024])
matching_teams = set(df_2024['CSAPATNÉV']).intersection(set(df_2022['CSAPATNÉV'])).intersection(set(df_2023['CSAPATNÉV']))
df = df_combined[df_combined['CSAPATNÉV'].isin(matching_teams)]

# Top 10 csapat kiválasztása
top_teams = df.groupby('CSAPATNÉV')['ÖSSZESEN'].sum().nlargest(10).index
df_top_teams = df[df['CSAPATNÉV'].isin(top_teams)]

last_10_teams = df.groupby('CSAPATNÉV')['ÖSSZESEN'].sum().nsmallest(10).index
df_last_10_teams = df[df['CSAPATNÉV'].isin(last_10_teams)]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
import matplotlib.pyplot as plt
import seaborn as sns

# Tanító és teszt adatok felosztása
X_train, X_test, y_train, y_test = train_test_split(df_top_teams[['ÉV']], df_top_teams['ÖSSZESEN'], test_size=0.2, random_state=42)

# Döntési fa regresszor létrehozása és tanítása
model = DecisionTreeRegressor()
model.fit(X_train, y_train)

# Predikciók készítése a teszt adatokra
predictions = model.predict(X_test)

# Diagram készítése
plt.figure(figsize=(12, 8))
sns.lineplot(x='ÉV', y='ÖSSZESEN', hue='CSAPATNÉV', data=df_top_teams, marker='o')
plt.plot(X_test, predictions, marker='o', linestyle='--', color='black', label='Predictions')
plt.title('Top 10 csapat becsült és valós teljesítménye az évek során')
plt.xlabel('Év')
plt.ylabel('Összesített Pontszámok')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
